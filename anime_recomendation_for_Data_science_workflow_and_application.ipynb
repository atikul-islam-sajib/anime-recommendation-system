Changing the character (,) to the character (|) in the genre column" in English

```python
anime['genre'] = anime['genre'].str.replace(',',' | ')
anime
```

Sorting anime based on anime_id and then storing it in the variable fix_anime

```python
fix_anime = anime.sort_values('anime_id', ascending=True)
fix_anime
```

Checking the number of fix_anime.

```python
len(fix_anime.anime_id.unique())
```

Creating a variable named 'preparation' that contains the 'fix_anime' dataframe and then sorting it based on 'anime_id'.

```python
preparation = fix_anime
preparation.sort_values('anime_id')
```

Remove duplicate data using drop_duplicates() based on anime_id

```python
# Membuang data duplikat pada variabel preparation
preparation = preparation.drop_duplicates('anime_id')
preparation
```

Convert data series to list. In this case, using the tolist() function from the numpy library.

```python
anime_ID = preparation['anime_id'].tolist()
anime_name = preparation['name'].tolist()
anime_genre = preparation['genre'].tolist()
 
print(len(anime_ID))
print(len(anime_name))
print(len(anime_genre))
```

Create a dictionary to determine key-value pairs in the previously prepared movie_id, movie_name, and movie_genre data.

```python
anime_new = pd.DataFrame({
    'id': anime_ID,
    'anime_name': anime_name,
    'genre': anime_genre
})
anime_new
```

Modeling

The tfidfvectorizer() function from the sklearn library

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tf = TfidfVectorizer()
tf.fit(anime_new['genre']) 
```

Perform fit and transformation into matrix form.

```python
tfidf_matrix = tf.fit_transform(anime_new['genre']) 
tfidf_matrix.shape 
```

Generate the tf-idf vector in matrix form, using the todense() function.

```python
tfidf_matrix.todense()
```

```python
pd.DataFrame(
    tfidf_matrix.todense(), 
    columns=tf.get_feature_names(),
    index=anime_new.anime_name
).sample(22, axis=1).sample(10, axis=0)
```

 **Cosine Similarity**

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(tfidf_matrix) 
cosine_sim
```

View the similarity matrix for each anime by displaying the anime title in 5 sample columns

```python
cosine_sim_df = pd.DataFrame(cosine_sim, index=anime_new['anime_name'], columns=anime_new['anime_name'])
print('Shape:', cosine_sim_df.shape)
 
cosine_sim_df.sample(5, axis=1).sample(10, axis=0)
```

Create an anime_recommendations function with the following parameters:

- anime_name : The title name of the movie (dataframe similarity index).
- Similarity_data : Dataframe regarding similarity that we have defined before
- Items : Names and features used to define similarities, in this case are 'anime_name' and 'genre'.
- k : Many recommendations to be given.




```python
def anime_recommendations(nama_anime, similarity_data=cosine_sim_df, items=anime_new[['anime_name', 'genre']], k=5):
   
    index = similarity_data.loc[:,nama_anime].to_numpy().argpartition(
        range(-1, -k, -1))
    closest = similarity_data.columns[index[-1:-(k+2):-1]]
    
    closest = closest.drop(nama_anime, errors='ignore')
 
    return pd.DataFrame(closest).merge(items).head(k)
```

Apply the code above to find anime recommendations that are similar to the Gintama anime.

```python
anime_new[anime_new.anime_name.eq('Gintama')]
```

Get anime recommendations similar to Gintama.

```python
anime_recommendations('Gintama')
```

## b- About Dataset::::
Context

This data set contains information on user preference data from 73,516 users on 12,294 anime. Each user is able to add anime to their completed list and give it a rating and this data set is a compilation of those ratings.

Content

Anime.csv

anime_id - myanimelist.net's unique id identifying an anime.
name - full name of anime.
genre - comma separated list of genres for this anime.
type - movie, TV, OVA, etc.
episodes - how many episodes in this show. (1 if movie).
rating - average rating out of 10 for this anime.
members - number of community members that are in this anime's
"group".

Rating.csv

user_id - non identifiable randomly generated user id.
anime_id - the anime that this user has rated.
rating - rating out of 10 this user has assigned (-1 if the user watched it but didn't assign a rating).
Acknowledgements
Thanks to myanimelist.net API for providing anime data and user ratings.

Building a better anime recommendation system based only on user viewing history.
